# Survey Data visualization for the plots used in slides

```python
import seaborn as sns
from matplotlib.pyplot import figure
import pandas as pd
df = pd.read_csv('China_2012_full_ES_N2700_data.csv', encoding='cp1252')
```

```python
k_columns = df.iloc[:, 176:213]
```

```python
k_columns.head(5)
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

explode = (0.05, 0, 0, 0, 0, 0)
colors = ['#F66D44', '#FEAE65','#E6F69D','#AADEA7', '#64C2A6', '#2D87BB']
total_frac = df['k3a'].value_counts()

total_frac_df = pd.DataFrame(
    data = {'fraction': total_frac.index, 'value' :total_frac.values})

top_frac = total_frac_df[:5].copy()

others = pd.DataFrame(data = {
    'fraction' : ['other'],
    'value' : [total_frac_df['value'][5:].sum()]
})

top_frac = pd.concat([top_frac, others])

labels = top_frac['fraction']
sizes = top_frac['value']

fig1, ax1 = plt.subplots(figsize = (16,10))
ax1.pie(sizes, labels=labels, autopct='%1.2f%%', explode=explode, colors=colors,
        shadow=True, startangle=90, textprops={'fontsize': 14})
plt.figtext(.5,.9,'Internal Funds/Retained earnings', fontsize=20, ha='center')
ax1.axis('equal');
```
